# Probability Distributions


## Continuos Distributions

Represented by PDA (Probability Density Functions).

### Uniform

A variable is known to lie in an interval and it's equally likely to take any value inside that interval. 

```python
import numpy as np
import seaborn as sns

nd = np.random.uniform(0,1,100)
sns.distplot(nd)
```

### Normal

Sample averages are normally distributed by the CLT (Central Limit Theory). 

- Addition: the sum of two independent normally distribuited variables is also normal.
- Mixture: If PD(X/Y) is normal and Y is normal X is also normal. This is useful for hierarchical models.

```python
import numpy as np
import seaborn as sns

nd = np.random.normal(0,1,100)
sns.distplot(nd)
```

## Discrete Distributions

Represented by PMA (Probability Mass Functions).

### Poisson

Use to represent count data (independent events in a period of time) which is useful to predict how many times an event can occur in a given time period.

```python
import numpy as np

# Probability of k events occurring within the interval lambda : l.
l1 = 1
pd1 = np.random.poisson(l1, size=10)

l2 = 5
pd2 = np.random.poisson(l2, size=10)

sns.distplot(pd1, label=str(l1))
sns.distplot(pd2, label=str(l2))
```

### Binomial

Represent number of successes in a sequence of Bernoulli trials.

```python
n1, p1 = 10, .2  # number of trials, probability of success
bd1 = np.random.binomial(n1, p1, 100)

# bd2 has higher success probability
n2, p2 = 10, .5  # number of trials, probability of success
bd2 = np.random.binomial(n2, p2, 100)

sns.distplot(bd1, label=str(p1))
sns.distplot(bd2, label=str(p2))
```
